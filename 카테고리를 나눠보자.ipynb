```python
df = pd.read_csv('../Data/df_유통사.csv').sort_values(by='고객번호')
```

```python
test1 = df[['고객번호', '대분류']]
test1.head()
```

```python
test1_2 = test1.groupby(['고객번호', '대분류'], as_index=False).size()
test1_2 = test1_2.sort_values(by=['고객번호', 'size'], ascending=False)
```

```python
test1_2['rank'] = test1_2.groupby("고객번호")["size"].rank("dense", ascending=False)
```

```python
test1_rank = test1_2[test1_2['rank']<=3]
test1_rank.head()
```

```python
test1_rank.drop(['size', 'rank'], axis=1, inplace=True)
```

```python
test1_rank
```

```python
test1_top3 = test1_rank.set_index('고객번호')
```

```python
from tqdm import tqdm
keys=list(set(test1_top3.index))
values = []
for i in tqdm(keys):
    values.append(test1_top3.loc[i].values.squeeze())
values
```

```python
top3 = pd.DataFrame({'고객번호':keys, '품목':values})
```

```python
top3['품목'] = top3['품목'].apply(lambda x: ', '.join(sorted(list(map(x)))))
```
